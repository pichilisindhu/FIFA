```python
data=pd.read_csv("C:/datasets/FIFA.csv")
```

```python
data.head()
```

```python
#Renaming the column names
data.rename(columns={'Height':'Height (cm)',"Weight":"Weight (kg)",'A/W':'Attacking Work Rate',
                'D/W':'Defensive Work Rate','IR':'International Reputation','↓OVA':'Overall Rating ',
                'POT':'Potential',"FK Accuracy":"Free Kick Accuracy","W/F":"Wing Forward","SM":"Skill Moves"},inplace=True)
```

```python
##removing the newline characters in club column
data['Club']=data['Club'].str.strip()
```

```python
##Removing the "cm","Kg" from height and weight columns
data["Height (cm)"]=data["Height (cm)"].str.rstrip("cm")
data["Weight (kg)"]=data["Weight (kg)"].str.rstrip("kg")
```

```python
# data[""]=data["Wing Forward"].str.rstrip("★")
data[['Wing Forward', 'Skill Moves','International Reputation']] = data[['Wing Forward', 'Skill Moves','International Reputation']].apply(lambda x: x.str.strip("★"))
```

```python
data['Value'] = data['Value'].str.replace('€','').str.strip()
data['Release Clause'] = data['Release Clause'].str.replace('€','').str.strip()
data['Wage'] = data['Wage'].str.replace('€','').str.strip()
```

```python
data["Value"].unique()
```

```python
convert_column=["Value","Wage","Release Clause"]
```

```python
def Value_change(value):
    if "M" in value: 
        return float(value.replace('M',''))
    elif "K" in value:
        return float(value.replace("K",''))/1000
    return value
    
```

```python
for x in convert_column:
    data[x]=data[x].apply(Value_change)
    
```

```python
data["Height (cm)"].unique()
```

```python
data["Height (cm)"].dtype
```

```python
def convert(value):
     if '\'' in value :
        feet, inches = value.split("'")
        inches = inches.strip('"')
        feet = int(feet)
        inches = int(inches)
        total_inches = feet * 12 + inches
        return (total_inches * 2.54)
    
     else:
        return value
     
```

```python
data["Weight (kg)"].unique()
```

```python
data["Height (cm)"]=data["Height (cm)"].apply(convert)
```

```python
def convert_weight(value):
    if 'lbs' in value:
        x=int(value.replace("lbs",""))
        x*=0.45
        return x
    else:
        return value
```

```python
data["Weight (kg)"]=data["Weight (kg)"].apply(convert_weight)
```

```python
data["Weight (kg)"]=data["Weight (kg)"].astype("int64")
```

```python
data[["contract start","contract end"]]=data["Contract"].str.split("~",expand=True)
```

```python
# Remove '\n' on the column
import re
data['contract start'] = data['contract start'].str.replace('\n','').str.strip()

# Extract the date and update the 'loan date end' column
data['Loan Date End'] = data['Club'].str.extract(r'(\w+ \d{1,2}, \d{4})')

# Function to remove last 20 characters using regex
def trim_last_n_chars(s, n):
    pattern = rf'.{{{n}}}$'
    return re.sub(pattern, '', s)

# Apply the function to the 'Team_Info' column
data['Club'] = data['Club'].apply(lambda x: trim_last_n_chars(x, 20))

# Check the shape of ['contract start','contract end']
print(data['contract start'].unique())
print("-"*100)
print(data['contract end'].unique())
```

```python
data['contract start'] = data['contract start'].str.replace(r' On Loan.*', '', regex=True).str.strip()
```

```python
print(data['contract start'].unique())
print("-"*100)
print(data['contract end'].unique())
```

```python
data.drop("Contract",axis=1,inplace=True)
```

```python
data.head()
```

```python
# Data types converson for various columns

col1 = ['Loan Date End','Joined','contract start','contract end'] # date

col2 = ['Wing Forward','Skill Moves','International Reputation','Hits'] #int


for col in col1:
    data[col] = pd.to_datetime(data[col],errors='coerce')

for col in col2:
    data[col] = pd.to_numeric(data[col],errors='coerce').astype('Int64')
```

```python

data = data.drop_duplicates()

data.duplicated().sum()
```

```python
# Impute contract start with the date of joining the club
data['contract start'] = data['contract start'].fillna(data['Joined'])

# Impute hits with its mean
data['Hits'] = data['Hits'].fillna(data['Hits']).mean() #loan date end	


# Impute contract_end with the date the player's loan ended
data['contract end'] = data['contract end'].fillna(data['Loan Date End'])
```
